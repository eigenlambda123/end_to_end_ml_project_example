```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Linear Regression

# Data
data_root = "https://github.com/ageron/data/raw/main/lifesat/"
lifesat = pd.read_csv(data_root + "lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Visualize Data
lifesat.plot(kind='scatter', x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.grid(True)
plt.show()

# Select Model
model = LinearRegression()

# Train Model
model.fit(X, y)

# Prediction
X_new = [[37_655.2]]  # Example input
print(model.predict(X_new))
```

```python
# k-nearest neighbors

# Data
data_root = "https://github.com/ageron/data/raw/main/lifesat/"
lifesat = pd.read_csv(data_root + "lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

# Visualize Data
lifesat.plot(kind='scatter', x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.grid(True)
plt.show()

# Select Model
model = KNeighborsRegressor(n_neighbors=3)

# Train Model
model.fit(X, y)

# Prediction
X_new = [[37_655.2]]  # Example input
print(model.predict(X_new))
```
